```python
def column(matrix,index):
  return [row[index] for row in matrix]

def swap_row(matrix, base_index, swapping_index):
  base_vals = matrix[base_index].copy()
  swap_vals = matrix[swapping_index].copy()
  matrix[base_index] = swap_vals
  matrix[swapping_index] = base_vals
  return matrix

def arg_max(matrix):
  max_val = 0
  index = -1
  for _ in range(len(matrix)):
    if matrix[_] > max_val:
      max_val = matrix[_]
      index = _
  return index

def num_of_cols(matrix):
  return len(matrix[0])

def multiply(matrix, number):
  a = [number*element for element in matrix]
  return a

def subtract(vec, vec2):
  a = [vec[_] - vec2[_] for _ in range(len(vec))]
  return a


def gauss_elim(matrix, output):
  base_cols = num_of_cols(matrix)
  augmented_matrix = [matrix[i] + [output[i]] for i in range(len(output))]

  for _ in range(base_cols):
  #  swap_index = arg_max(column(augmented_matrix,_))
  #  if not swap_index == _:
  #    augmented_matrix = swap_row(augmented_matrix,_,swap_index)
    for i in range(_+1,len(augmented_matrix)):
      if not column(augmented_matrix,_)[_] == 0:
        factor = column(augmented_matrix,_)[i] / column(augmented_matrix,_)[_]
      
        augmented_matrix[i] = subtract(augmented_matrix[i], multiply(augmented_matrix[_],factor))


  for _ in range(base_cols):
    for i in range(_+1,len(augmented_matrix)):
      if not column(augmented_matrix,base_cols -1  - _)[base_cols -1- _] == 0:
        factor = column(augmented_matrix,base_cols -1  - _)[len(augmented_matrix)-1-i] / column(augmented_matrix,base_cols -1  - _)[base_cols -1- _]
        augmented_matrix[len(augmented_matrix) - 1 -i] = subtract(augmented_matrix[len(augmented_matrix)-1-i], multiply(augmented_matrix[base_cols -1- _],factor))
  
  x = []
  for _ in range(len(augmented_matrix)):
    if augmented_matrix[_][_] == 0:
      x.append(augmented_matrix[_][_])
    else:
      x.append(augmented_matrix[_][-1]/ augmented_matrix[_][_])
  return x, augmented_matrix
```
